```python
import pandas as pd

vac_csv = './export_data/VacBldg.csv'

vac_data = pd.read_csv(vac_csv)

addr_data = vac_data[['SurveyYear','AddrNum','StName','AddrSuf']]
addr_data=addr_data.sort_values(by=['AddrSuf','StName','AddrNum'])

addr_data['Years Vacant']=0
addr_data['Vacancy Start']=0
addr_data['Vacancy End']=0

curr_addr=''
curr_st=''
curr_suf=''
startyear=0
endyear=0
years=0

addr_data=addr_data.reset_index(drop=True)

#addr_data=addr_data[0:100]

repeat=False


for index,row in addr_data.iterrows():
    if(index==0):
        curr_addr=row['AddrNum']
        curr_st=row['StName']
        startyear=row['SurveyYear']
    else:
        if((curr_addr!=row['AddrNum'] or curr_st!=row['StName'] or curr_suf!=row['AddrSuf']) and repeat==True):
            addr_data.loc[index-1, 'Vacancy Start']=startyear
            addr_data.loc[index-1, 'Vacancy End']=endyear
            addr_data.loc[index-1, 'Years Vacant']=years
            curr_addr=row['AddrNum']
            curr_st=row['StName']
            curr_suf=row['AddrSuf']
            startyear=row['SurveyYear']
            repeat=False
            years=0
        elif((curr_addr!=row['AddrNum'] or curr_st!=row['StName'] or curr_suf!=row['AddrSuf']) and repeat==False):
            addr_data.loc[index-1, 'Vacancy Start']=startyear
            addr_data.loc[index-1, 'Vacancy End']=startyear
            curr_addr=row['AddrNum']
            curr_st=row['StName']
            curr_suf=row['AddrSuf']
            startyear=row['SurveyYear']
            repeat=False
            years=0
            
        else:
            endyear=row['SurveyYear']
            years+=1
            repeat=True

addr_data=addr_data.drop(addr_data[addr_data['Vacancy Start']==0].index)

morethanzero=addr_data.drop(addr_data[addr_data['Years Vacant']==0].index)


addr_data
```

```python
print("Number of buildings vacant for less than a year",len(addr_data)-len(morethanzero))
print("Number of buildings vacant for more than a year",len(morethanzero))
print("Average number of years a building is vacant: ",round(addr_data['Years Vacant'].mean(),2))
print("Average of buildings vacant for more than a year: ",round(morethanzero['Years Vacant'].mean(),2))
for i in range(0,(2018-1989)+1):
    print("Number of buildings vacant for ",i," years:",len(addr_data.drop(addr_data[addr_data['Years Vacant']!=i].index)))
```
